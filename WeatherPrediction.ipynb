```python
trained_model = YOLO('/content/runs/classify/train2/weights/best.pt')
```

```python
from matplotlib import pyplot as plt
import cv2
def detectWeather(image_path):
  results = trained_model.predict(image_path)
  for result in results:
        img = result.plot()  # Draw bounding boxes
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
```
